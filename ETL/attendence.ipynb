Load to postgres db

```python
import pandas as pd
from sqlalchemy import create_engine
import os



# PostgreSQL connection parameters
pg_dbname =     os.getenv('pg_dbname')
pg_user =       os.getenv('pg_user')
pg_password =   os.getenv('pg_password')
pg_host =       os.getenv('pg_host')
pg_port =       os.getenv('pg_port')  # Default PostgreSQL port

print(pg_dbname)
print(pg_user)
print(pg_password)
print(pg_host)
print(pg_port)


# Create SQLAlchemy engine


engine = create_engine(f'postgresql://{pg_user}:{pg_password}@{pg_host}:{pg_port}/{pg_dbname}')


# Dump DataFrame into PostgreSQL using SQLAlchemy engine
table_name = 'employees'
df.to_sql(table_name, engine, if_exists='replace', index=False)

table_name = 'attendence'
df1.to_sql(table_name, engine, if_exists='replace', index=False)


print("Data loaded into PostgreSQL table successfully!")
```
